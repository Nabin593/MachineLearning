```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=0
lr=0.1
error=[]
for i in range(50):
  grad = 2*(x-5)
  x = x-lr*grad
  y = (x-5)**2
  error.append(y)
  plt.scatter(x,y)
  print(x)
plt.plot(error)
plt.show()
```
