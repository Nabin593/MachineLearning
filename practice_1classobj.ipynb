```python
class student :
  def __init__(self, name, age) :
    self.name = name
    self.age = age
  def studfunc(self) :
    print("My name is : " + self.name)

a1 = student("Ram", 15)
a1.studfunc()
```

```python
class car :

  # defining a constructor
  def __init__(self, name, fuel) :
    self.carname = name
    self.fuelType = fuel

    # defining a function
  def printfunc(self) :
    print(self.carname, self.fuelType)

# passing value to the class
p1 = car("alto", "petrol")
p1.printfunc()
```
